```python
import glob
import matplotlib.pyplot as plt
import cv2
```

```python
# read data
def read_images(img_dir):
    images = [cv2.imread(file, 0) for file in glob.glob(img_dir+"/*.png")]
    return images
ims = read_images('imgs')
```

```python
# plot all images
fig, axs = plt.subplots(4,4, figsize=(6, 6))
for im_idx, im in enumerate(ims):
    row = im_idx // 4
    col = im_idx % 4
    axs[row, col].imshow(ims[im_idx], 'gray')
    axs[row, col].set_title(f'im {im_idx + 1}')
    axs[row, col].set_xticks([])
    axs[row, col].set_yticks([])
fig.tight_layout()
```

```python
# copy image sets from matlab script
set1 = [5,14,9,7];
set2 = [5,3,15,1];
set3 = [5,3,4,1];
set4 = [5,11,4,1];
```

```python
# plot image set
sets = set1 + set2 + set3 + set4
fig, axs = plt.subplots(4,4, figsize=(6, 6))
for idx, im_idx in enumerate(sets):
    row = idx // 4
    col = idx % 4
    axs[row, col].imshow(ims[im_idx-1], 'gray')
    axs[row, col].set_title(f'im {im_idx}')
    axs[row, col].set_xticks([])
    axs[row, col].set_yticks([])
fig.tight_layout()
```

```python
# plot image set ready for figure
sets = set1 + set2 + set3 + set4
coloured_rows = [0, 1, 2, 3, 0, 1, 2, 0, 1, 0]
coloured_cols = [0, 0, 0, 0, 1, 1, 1, 2, 2, 3]
col_labels = ['A', 'B', 'C', 'D']
fig, axs = plt.subplots(4,4, figsize=(6, 6))
for idx, im_idx in enumerate(sets):
    row = idx // 4
    col = idx % 4
    axs[row, col].imshow(ims[im_idx-1], 'gray')
    axs[row, col].set_xticks([])
    axs[row, col].set_yticks([])
for row, col in zip(coloured_rows, coloured_cols):
    for spine in axs[row, col].spines.values():
        spine.set_edgecolor('orange')
for row in range(4):
    for col in range(4):
        for spine in axs[row, col].spines.values():
            spine.set_linewidth(3)
for col, letter in enumerate(col_labels):
    axs[3, col].set_title(letter, size=40, pad=30, y=-1)
fig.tight_layout()
fig.savefig("image_set.eps")
```
